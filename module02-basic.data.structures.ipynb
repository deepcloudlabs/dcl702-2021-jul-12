```python
# generator function + functional programming
numbers = [4,8,15,16,23,42]
```

```python
# problem: sum( even ** 2 )
total = 0
for number in numbers:
  if number%2 == 0: # 4 8 16 42
    squared = number ** 2
    total += squared # 16 64 256 1764
print(total)    
```

```python
def even(num):
  print(f"even({num}) is working....")
  return num%2 == 0
```

```python
def take_square(n):
  print(f"take_square({n}) is working....")
  return n**2
```

```python
sum(map(take_square, filter( even , numbers))) # pipeline -> function -> no side-effect/high-order/pure function -> lambda function, 1-liner
```

```python
# filter / map /reduce : sum
sum(map(lambda u : u ** 2, filter( lambda num : num%2 == 0 , numbers))) # pipeline -> function -> no side-effect/high-order/pure function -> lambda function, 1-liner
```

```python
def classical_3n_plus_1(n):
  nums = []
  while n > 1:
    if n%2 == 0:
      n = n // 2
    else: 
      n = 3 * n + 1
    nums.append(n)
  return nums   
```

```python
for i in classical_3n_plus_1(17):
  print(i)
```

```python
# 17 -> 52 -> 26 -> 13 -> 40 -> 20 -> 10 -> 5 -> 16 -> 8 -> 4 -> 2 -> 1
def _3n_plus_1(n): # generator
  while n > 1:
    if n%2 == 0:
      n = n // 2
    else: 
      n = 3 * n + 1
    yield n
```

```python
result = _3n_plus_1(17)
```

```python
next(result)
```

```python
next(result)
```

```python
next(result)
```

```python
for i in _3n_plus_1(17):
  print(i)
```

```python
# list, set, tuple, dictionary -> memory -> files
# 1. text i/o    : str -> files -> text editor -> HRF, xml/json/svg/csv/... -> Human
# 2. binary i/o  : memory -> byte array -> files, mp3, mp4, jpeg, png, bson -> A2A Integration
employees = [
    ("jack shephard", "Sales", 100000, 1978, True),
    ("kate austen", "IT", 200000, 1985, True),
    ("ben linus", "Finance", 150000, 1967, True),
    ("james sawyer", "HR", 70000, 1970, True),
    ("kim kwon", "Sales", 120000, 1986, True),
    ("sun kwon", "IT", 200000, 1969, False),
    ("hugo reyes", "IT", 120000, 1992, True)
]
```

```python
try:
  with open("employees.csv", mode="w") as my_file:
    for full_name,dept,salary,birth_year,full_time in employees:
      my_file.write(f"{full_name.title()},{dept.upper()},{salary},{birth_year},{'FULL_TIME' if full_time else 'PART_TIME'}\n")
except PermissionError:
  print("Cannot open/write to file. Try to change permission!")
except Exception as err:
  print("An error has occurred: "+str(err))        
```
