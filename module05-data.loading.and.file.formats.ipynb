```python
df = pd.read_csv("sample_data/eurostat_hicpv2.csv")
```

```python
eu_inflations = df.iloc[6:]
```

```python
eu_inflations.info()
```

```python
eu_inflations
```

```python
years = eu_inflations.columns.to_list()[1:]
```

```python
inflations = pd.melt(frame=eu_inflations, id_vars='geo', value_vars=years, var_name="years", value_name="inflation_rate")
```

```python
fig = px.line(inflations, x="years", y="inflation_rate", color="geo", title="Inflation Rates Comparison")
fig.show()
```
