```python
meyveler = ["    elma    ", "\t\tarmut\t\n\n\n", "         kiraz\n\n\n", "karpuz", "muz", "seftali", "kivi", "ananas"]
```

```python
fruits
```

```python
import re
```

```python
pattern1 = r"^k[a-z]*z$" # ., *, +, ?, ^, $ , {}, [^ai]
pattern2 = r"^[a-z]{6,}$"
pattern3 = r"^[^ai]+$"
reg1 = re.compile(pattern3, flags= re.IGNORECASE)
```

```python
for fruit in fruits:
  if reg1.match(fruit):
    print(fruit)
```
