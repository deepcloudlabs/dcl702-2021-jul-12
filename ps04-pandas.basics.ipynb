```python
# Exercise 1 : employees -> dictionary -> DataFrame
# full_name: str, salary: float, department: str, birth_year: int, full_time: True ? False
```

```python
import pandas as pd
import numpy as np
```

```python
employees = {
    "full_name": ["jack shephard", "kate austen", "james sawyer", "ben linux", "hugo reyes"],
    "salary": [100_000., 200_000., 150_000., 250_000., 300_000.],
    "department": ["IT", "SALES", "IT", "SALES", "FINANCE"],
    "birth_year": [1976, 1983, 1978, 1956, 1991],
    "full_time": [True, False, True, False, True]
}
```

```python
df_emps = pd.DataFrame(employees)
```

```python
df_emps
```

```python
# exercise 2: add new column "part_time"
df_emps["part_time"] = ~df_emps["full_time"] # logical not operator: ~
```

```python
# exercise 3: get full_time employees in a DataFrame
df_emps[df_emps.full_time]
```

```python
# exercise 4: sort full_time employees by full_name
df1 = df_emps[df_emps.full_time].sort_values(by=["birth_year"],ascending=True)
```
