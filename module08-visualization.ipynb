```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
fig = plt.figure()
sp1 = fig.add_subplot(2,2,1)
sp2 = fig.add_subplot(2,2,2)
sp3 = fig.add_subplot(2,2,3)
sp4 = fig.add_subplot(2,2,4)
sp1.hist(np.random.randn(2000), bins=20, color='g', alpha=0.1)
sp2.scatter(np.arange(30)+ 2 * np.random.randn(30),np.arange(30) + 3 * np.random.randn(30))
sp3.plot(np.random.randn(50).cumsum(), 'r--')
sp4.plot(np.arange(30))
fig.show()
```
